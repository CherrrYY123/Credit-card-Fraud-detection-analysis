# Credit Card Fraud Detection Analysis

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score  
```

```python
CardData = pd.read_csv('creditcard.csv')
```

```python
CardData.head()
```

```python
CardData.tail()
```

```python
CardData.info()
```

```python
CardData.isnull().sum()
```

```python
CardData['Class'].value_counts()
#1- Fraudulent Transactions
#0- Legal Transactions
```

```python
##Seperating the data Classwise
legal = CardData[CardData.Class == 0]
fraud = CardData[CardData.Class == 1]
```

```python
print(legal.shape)
```

```python
print(fraud.shape)
```

```python
legal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
CardData.groupby('Class').mean() #Mean of all the columns
```

```python
#Creating Sample datasets of legal and fraudulent transactions: UNDERSAMPLING
#To get similar distributions of both classes of transactions
```

```python
legal_sample = legal.sample(n = 492)
```

```python
NewData = pd.concat([legal_sample,fraud], axis = 0)
```

```python
NewData.head()
```

```python
NewData.tail()
```

```python
NewData['Class'].value_counts()
```

```python
NewData.groupby('Class').mean()
```

```python
X = NewData.drop(columns = 'Class', axis = 1)
```

```python
print(X)
```

```python
Y = NewData['Class']
print(Y)
```

```python
#Splitting the above randomly into Training and Testing Data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
#Data stored in X_train and their corresponding labels in Y_train which are simply 0 or 1, simlilarly for X_test and Y_test
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Logistic Regression
```

```python
model = LogisticRegression(solver='liblinear', class_weight='balanced', max_iter=100)
```

```python
model.fit(X_train, Y_train) #Training the model with Training data
```

```python
#Predicting Accuracy of the model
```

```python
#Training data Accuracy
X_trainPredict = model.predict(X_train)
Accuracy = accuracy_score(X_trainPredict, Y_train)
print("Accuracy for training data:" , Accuracy)
```

```python
#Testing data Accuracy
X_testPredict = model.predict(X_test)
test_accuracy = accuracy_score(X_testPredict, Y_test)
print("Accuracy for testing data:" , test_accuracy )
```
